```python
df_caracteristicas = df_caracteristicas.fillna(0)

df_caracteristicas.isna().sum()
```

```python
df_caracteristicas['work_rate'].unique()
```

```python
df_caracteristicas[['attacking_work_rate', 'defending_work_rate']] = df_caracteristicas['work_rate'].str.split('/', expand=True)
df_caracteristicas.drop(columns='work_rate', axis=1, inplace=True)
df_caracteristicas.head()
```

```python
# Para manter na escala de 0 - 100, vamos atribuir valores numéricos à essas duas colunas que podem ser traduzidas como 
# Comprometimetno no ataque e Comprometimento na defesa. Já que não há diferenciação baseado no overall, 33 será baixo comprometimento, 66 médio e 99 alto

df_caracteristicas['attacking_work_rate'] = df_caracteristicas['attacking_work_rate'].apply(lambda x: 33 if x == 'Low' else (66 if x == 'Medium' else 99))


df_caracteristicas['defending_work_rate'] = df_caracteristicas['defending_work_rate'].apply(lambda x: 33 if x == 'Low' else (66 if x == 'Medium' else 99))
```

```python
df_caracteristicas.head()
```

```python
# Para facilitar as recomendações, iremos agregar algumas posições para diminuir a quantidade de posições diferentes

traducao_posicoes = {'CF': 'Atacante',
 'RW': "Extremo",
 'ST': "Atacante",
 'CAM': "Meio campo",
 'RM': "Meio campo",
 'CM': "Meio campo",
 'LW': "Extremo",
 'CB': "Zagueiro",
 'GK': "Goleiro",
 'CDM': "Volante",
 'LB': "Lateral",
 'LM': "Meio campo",
 'RB': "Lateral",
 'RWB': "Ala",
 'LWB': "Ala"}
```

```python
# Traduzindo as abreviações e removendo repetições
df_caracteristicas['player_positions'] = df_caracteristicas['player_positions'].str.split(',').apply(
    lambda posicoes: list(set([traducao_posicoes[pos.strip()] for pos in posicoes])) if isinstance(posicoes, list) else posicoes)

# Convertendo de volta para string, com as posições separadas por vírgulas
df_caracteristicas['player_positions'] = df_caracteristicas['player_positions'].apply(
    lambda posicoes: ', '.join(sorted(posicoes)) if isinstance(posicoes, list) else posicoes)
```
