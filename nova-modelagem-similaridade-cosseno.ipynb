```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("jogadores-por-posicao\\todos_os_jogadores.csv")
```

```python
features = df.columns.difference(['player_id', 'short_name', 'long_name', 'player_positions', 'overall'])
```

```python
# Calcular a matriz de similaridade de cosseno
similaridades = cosine_similarity(df[features])
```

```python
df.head(30)
```

```python
# Criar um DataFrame para exibir as similaridades entre os jogadores


df_similaridades = pd.DataFrame(similaridades, index=df['short_name'], columns=df['short_name'])

print("\nMatriz de Similaridade (quanto maior, mais parecido):")

df_similaridades.head(40)
```

```python
# Ordenar jogadores mais semelhantes ao jogador de referência (excluindo ele mesmo)
jogador_ref = 'L. Hernández'
recomendados = df_similaridades[jogador_ref].sort_values(ascending=False)[1:6]  # Excluir o próprio jogador

# Criar um DataFrame com os jogadores recomendados
df_recomendados = df[df['short_name'].isin(recomendados.index)][['short_name', 'overall']]

# Adicionar a coluna de similaridade ao DataFrame
df_recomendados['similaridade'] = recomendados.values

# Exibir o DataFrame final
print("\nJogadores recomendados para {}:".format(jogador_ref))
print(df_recomendados)
```

```python
# Obter os dados normalizados do jogador de referência
dados_ref = df[df['short_name'] == jogador_ref][features].iloc[0]

# Obter os dados dos jogadores recomendados
jogadores_recomendados = df[df['short_name'].isin(recomendados.index)][features]

# Criar o DataFrame de comparação, incluindo apenas o jogador de referência e os recomendados
jogadores_comparacao = pd.concat([dados_ref.to_frame().T, jogadores_recomendados])

# Criar a lista de nomes de jogadores para a legenda (não incluir o jogador de referência aqui)
nomes_jogadores = list(recomendados.index)  # Apenas os recomendados para a legenda

# Criar o gráfico de radar
num_features = len(features)
angles = np.linspace(0, 2 * np.pi, num_features, endpoint=False).tolist()
angles += angles[:1]  # Fechar o círculo (adicionando o primeiro valor de volta ao final)

# Criar o gráfico com tamanho ajustado
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

# Plotar cada jogador no gráfico de radar
for nome, row in jogadores_comparacao.iterrows():
    valores = row.tolist()
    
    # Adicionar o primeiro valor ao final para "fechar" o gráfico
    valores += valores[:1]
    
    # Usar o nome do jogador para a legenda
    ax.plot(angles, valores, linewidth=2, label=nome)  # Usar o nome do jogador diretamente
    ax.fill(angles, valores, alpha=0.25)

# Ajustes finais do gráfico
ax.set_yticklabels([])  # Remover os rótulos do eixo Y
ax.set_xticks(angles[:-1])  # Evitar a última repetição do ângulo
ax.set_xticklabels(features, fontsize=10, rotation=45)  # Diminuir o tamanho da fonte

# Ajustar a posição da legenda para fora do gráfico
ax.legend(loc='upper left', bbox_to_anchor=(1.1, 1), fontsize=12)  # Ajustar a posição da legenda

# Título do gráfico
plt.title(f'Comparação de características: {jogador_ref} e recomendados', fontsize=16)

# Exibir o gráfico
plt.tight_layout()  # Ajuste automático do layout para evitar sobreposição
plt.show()
```
